```python
list_of_id = []

for i in range(1,11):
    url = "https://www.themoviedb.org/movie?page={}".format(i)
#     print(url)
    page = requests.get(url)

    soup = BeautifulSoup(page.content, 'html.parser')

    movies = soup.find(class_="results flex results_poster_card")

    movie_detail = movies.find_all(class_="flex")

    for i in range(len(movie_detail)):
        id_string = movie_detail[i].find('a')
        id_int = int(id_string['href'].split('/')[2])
        list_of_id.append(id_int)

```

```python
list_of_movie = []
for i in list_of_id:
    url = 'https://api.themoviedb.org/3/movie/{}?api_key=2f848bb1906d7277820e69e3fa9158a5'.format(i)
#     print(url)
    response = requests.get(url)

    data = response.json()
    keys = ['title','genres']
    parsed_results = {key: (data[key] if key in data else 'Null') for key in keys}
    list_of_movie.append(parsed_results)
    time.sleep(1)
    
```

```python
for movie in list_of_movie:
    
    for key, value in movie.items():

        kind_of_genres = []
        if key == 'genres':
            for i in value:
                kind_of_genres.append(i['name'])
                
            movie['genres'] = kind_of_genres
```

```python
movies_lists = []
for movie in list_of_movie:
    for i, gen in enumerate(movie['genres']):
        movie_tuple = (movie['title'], gen)
        movies_lists.append(movie_tuple)

movies_lists
```

```python
insert_stmt = "INSERT INTO The_Movie_Db_Genres ( movie_name, genres) VALUES (%s, %s)"
cursor.executemany(insert_stmt, movies_lists)
cnx.commit()
# time.sleep(1)
```
